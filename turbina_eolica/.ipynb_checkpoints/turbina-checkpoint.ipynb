```python
#Importando arquivos
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
#Lendo o arquivo
turbina = pd.read_csv('T1.csv')
display(turbina)
```

```python
#Mudando os nomes das columas
turbina.columns = ['Data/Hora', 'ActivePower(Kw)', 'WindSpeed(m/s)', 'CurvaTeórica(KWh)', 'DireçãoVento(º)']
display(turbina)
```

```python
#Deletando columa
del turbina['DireçãoVento(º)']
display(turbina)
```

```python
#Mudando o campo data para datetime
turbina['Data/Hora'] = pd.to_datetime(turbina['Data/Hora'])
display(turbina)
```

```python
#Informações da base de dadps
turbina.info()
```

```python
#Plotando os dados em um gráfico - Real
sns.scatterplot(data=turbina, x="WindSpeed(m/s)", y="ActivePower(Kw)")
```

```python
#Plotando os dados em um gráfico - Real
sns.scatterplot(data=turbina, x="WindSpeed(m/s)", y="CurvaTeórica(KWh)")
```
