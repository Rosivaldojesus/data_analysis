```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Abrindo o aquivo CSV
df = pd.read_csv("lista_plantas.csv", encoding= 'unicode_escape', sep=";")
df.head()
```

```python
#Verificando as informações do dataframe
df.info()
```

```python
# Verificando a quantidade por pais
pais = df.groupby(["Country"]).count()
pais
```

```python
#Ordenado do maior para o menor
df = pais.sort_values(by = "CustomerID", ascending=False)
df.head()
```

```python
#Resertando o indice
df.reset_index(inplace=True)
df
```

```python
# Gráfico
x = df["Country"]
y = df["CustomerID"]

plt.figure(figsize=(20,5))

plt.bar(x,y)
plt.show()
```
