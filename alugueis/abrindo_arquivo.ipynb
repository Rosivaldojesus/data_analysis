```python
import pandas as pd
import numpy as np
```

```python
# Criando um dataframe com lista/dicionários
alugueis = pd.read_csv("houses_to_rent.csv")
alugueis.head(2)
```

```python
alugueis["hoa_tratado"] = alugueis["hoa"].apply(lambda x: x.replace("R$", "").replace(",", ""))
```

```python
alugueis.head(2)
```

```python
# Utilizando uma biblioteca e a função rename
colunas = {"city": "cidade", "area": "area_m2", "bathroom": "banheiros"}
alugueis.rename(columns= colunas).head()
```

```python
# Utilizando uma biblioteca e a função rename
novos_nomes = ["indice", "cidade", "area_m2", "quartos", "banheiros", "vagas_estacionamneto", "andar", "aceito_animais", "mobiliado",
"hoa", "valor_aluguel", "taxas", "seguro_incendio", "total"]

colunas = dict((chave, valor) for chave, valor in zip(alugueis.columns.to_list(), novos_nomes))

alugueis.rename(columns= colunas).head()
```
